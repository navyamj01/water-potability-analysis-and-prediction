```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df_train = pd.read_csv("../data/train_processed.csv")
df_test = pd.read_csv("../data/test_processed.csv")
```

```python
X = df_train.drop("Potability", axis = 1)
y = df_train["Potability"]
```

```python
X_test = df_test.drop("Potability", axis = 1)
y_test = df_test["Potability"]
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(max_iter=1000)


log_model.fit(X, y)
```

```python
y_pred = log_model.predict(X_test)
```

```python
y_pred
```

This means that model learned that if it predicts 0 always then it is always correct
this will lead to two major reasons that:
1. log reg didn't worked for this
2. there is class imbalance

```python
y.value_counts()
# this confirms the class imbalnce issue, but it is not that imbalanced that it willlead to 100% prediction of 0
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000))
])

pipeline.fit(X, y)

y_pred = pipeline.predict(X_test)
```

```python
y_pred
```

```python
pipeline.predict_proba(X_test)[:10]
```

```python
probs = pipeline.predict_proba(X_test)[:, 1]

y_pred = (probs > 0.4).astype(int)

print(set(y_pred))
```

The model produced probabilities close to the decision boundary, so adjusting the classification threshold improved class balance.

```python
pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000, class_weight='balanced'))
])
pipeline.fit(X,y)
```

```python
probs = pipeline.predict_proba(X_test)[:, 1]
# this will give the prob of class 1
probs
```

```python
y_pred = pipeline.predict(X_test)
from collections import Counter

my_list = y_pred
counts = Counter(my_list)
```

```python
print(counts)
```

now this works well, but still it is not good as the line weight= balanced tell it to predict more 1's so it penalizes the wrong prediction more.
In real-world problems (like water safety):
1. Predicting unsafe water as safe is dangerous. So we intentionally make the model care more about that class
this is called cost learning

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
print("\n\n\n")
# well as 126 are FN, this is dangerous in water potability case so it is a little wrong or not fully right at predicting the potability

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
# in log reg, the best loss calculating technique is binary cross entropy in this case as results are only two : 0 and 1
from sklearn.metrics import log_loss

loss = log_loss(y_test, probs)
print("Log Loss:", loss)
```

```python
# this loss tells that this model is predicting randomly, hence it is not coorect for this specific problem
```

The obtained log loss value of approximately 0.69 indicates that the model's predicted probabilities are close to random guessing. 
This suggests that the model is unable to confidently distinguish between potable and non-potable water samples.

This behavior further confirms that Logistic Regression is underfitting the data and failing to capture underlying patterns effectively.
